# Import libraries

```python
from copy import deepcopy 
import numpy as np # linear algebra
from matplotlib import pyplot as plt #for plot
```

# create 3 centroids

```python
center_1 = np.array([1,1])
center_2 = np.array([5,6])
center_3 = np.array([9,1])
```

# generate random data around 3 centroids

```python
data_1 = np.random.randn(100,2) + center_1
data_2 = np.random.randn(100,2) + center_2
data_3 = np.random.randn(100,2) + center_3

data = np.concatenate((data_1, data_2, data_3), axis = 0)


plt.scatter(data[:,0], data[:,1], s=7)
```

# generate new point clusters


```python
n_cluster = 3
n_data = data.shape[0]
n_point = data.shape[1]

mean = np.mean(data, axis = 0)
std = np.std(data, axis = 0)
```

```python
# Generate random centers
centers = np.random.randn(n_cluster,n_point)* std + mean

# Plot the data and the centers generated as random
plt.scatter(data[:,0], data[:,1], s=7)
plt.scatter(centers[:,0], centers[:,1], marker='*', c='r', s=100)

```

# generate new centroids

```python
centers_old = np.zeros(centers.shape) # save for old centers
centers_new = deepcopy(centers) # Store new centers

data.shape
clusters = np.zeros(n_data)
distances = np.zeros((n_data,n_cluster))

dist = np.linalg.norm(centers_new - centers_old)
```

# compute distance between old and new centroids

# and energy

# Energy = 
# Sum of distance between point and centroid
# / number of clusters

```python

# When, after an update, the estimate of that center stays the same, exit loop
while dist != 0:
    # Measure the distance to every center
    i_energy = 0
    for i in range(n_cluster):
        distances[:,i] = np.linalg.norm(data - centers[i], axis=1)
        i_energy += distances[:,i]
    # Assign all data to closest center
    clusters = np.argmin(distances, axis = 1) 
    centers_old = deepcopy(centers_new)
    print('energy = ' ,i_energy / i, "\n", 
          '\ndata : ',data , '\nbetween center : ',
          centers[i] ,'\n') 
    # Energy = 
   # Sum of distance between point and centroid
   # / number of clusters
    # Calculate mean for every cluster and update the center
    for i in range(n_cluster):
        centers_new[i] = np.mean(data[clusters == i], axis=0)
    dist = np.linalg.norm(centers_new - centers_old)
    
```

# final optimized clustering

```python
# Plot the data and the centers generated as random
plt.scatter(data[:,0], data[:,1], s=7)
plt.scatter(centers_new[:,0], centers_new[:,1], marker='*', c='g', s=150)
```
